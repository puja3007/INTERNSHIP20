```python
url = "https://github.com/"
driver.get(url)
```

```python
WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary'))).click()
WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a'))).click()
```

```python
urls=[]
for i in driver.find_elements_by_xpath('//*[@id="js-pjax-container"]/div[3]/div/div[2]/article/h1/a'):
    urls.append(i.get_attribute("href"))
```

```python
Title=[]
Description=[]
contributer_count=[]
Language_used=[]
```

```python
for i in urls:
    driver.get(i)
    try:
        title=driver.find_element_by_xpath('//h1[@class=" d-flex flex-wrap flex-items-center break-word f3 text-normal"]')
        Title.append(title.text.replace('\n',' '))
    except NoSuchElementException:
        Title.append('-')
        
    try:
        description=driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[1]/div/p')
        Description.append(description.text)
    except NoSuchElementException:
        Description.append('-')
        
    try:
        count=driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[5]/div/h2/a/span')
        contributer_count.append(count.text)
    except NoSuchElementException:
        contributer_count.append('-')

    try:
        lang=driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[6]/div/ul/li/a/span[1]')
        Language_used.append(lang.text.replace('\n',' '))
    except NoSuchElementException:
        Language_used.append('-')
```

```python
df=pd.DataFrame({})
df['Repository_Title']=Title
df['Repository_Description']=Description
df['Contributer_Count']=contributer_count
df['Language_Used']=Language_used
df
```
